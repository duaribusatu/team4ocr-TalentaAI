```python
def to_document(row):
    DOCUMENT = {
        "id": str(row.name),
        "NamaPangan": row['Nama Pangan'],
        "NamaPangan_Emb": row['NamaPangan_Emb'],
        "Energi": row['Energi (Kal)'],
        "Protein": row['Protein (g)'],
        "Lemak": row['Lemak (g)'],
        "Karbohidrat": row['Karbohidrat (g)'],
        "Serat": row['Serat (g)'],
        "Kalsium": row['Kalsium (g)'],
        "Besi": row['Besi (g)'],
        "Garam": row['Garam (g)'],
        "VitaminA": row['Vitamin A (g)'],
        "VitaminB1": row['Vitamin B1 (g)'],
        "VitaminB2": row['Vitamin B2 (g)'],
        "VitaminB3": row['Vitamin B3 (g)'],
        "VitaminC": row['Vitamin C (g)'],
    }
    return DOCUMENT
documents = [to_document(row) for i, row in df.iterrows()]
len(documents)
```

# Search by Name

```python
def search_pangan_by_name(nama_pangan):
    results = search_client.search(search_text=nama_pangan)
    data = [result for result in results]
    df_res = pd.DataFrame(data)[['@search.score', 'Lemak', 'VitaminB1', 'VitaminB2', 'VitaminA', 'Besi', 'Karbohidrat', 'Serat', 'VitaminB3', 'id', 'NamaPangan', 'Energi', 'NamaPangan_Emb', 'VitaminC', 'Garam', 'Protein']]
    return df_res
```
