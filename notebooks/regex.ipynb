```python
def concat(row):
    cols = ['BAHAN', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']
    vals = row[cols].values
    vals = [str(val) for val in vals if pd.notna(val)]
    return unicodedata.normalize("NFKD", '\n'.join(vals))

df_resep['resep'] = df_resep.apply(lambda row: concat(row), axis=1)
df_resep_clean = df_resep[['Nama Makanan', 'resep']]
df_resep_clean
```

```python
def convert_to_grams(text):
    conversions = {
        "sdm": 12,
        "sdt": 4,
        "ml": 1,
        "suing": 5,
        "butir": 50,
        "buah": 150,
        "daun": 1,
        "potong": 100,
        "bh": 50,
        "biji": 50,
        "bj": 50
    }

    result = []
    pattern = r'(\d+)\s*(\S+)'

    for match in re.finditer(pattern, text):
        amount, unit = match.groups()
        amount = float(amount)
        if unit in conversions:
            grams = amount * conversions[unit]
            result.append((amount, unit, grams))

    return result

text = """
1 sdm = 12 gr
1 sdt = 4 gr
1 ml = 1 gr
1 suing = 5 gr
1 butir = 50 gr
1 buah = 150 gr
1 daun = 1 gr
1 potong = 100 gr
1 bh = 50 gr
1 biji = 50 gr
1 bj = 50 gr
"""

conversions = convert_to_grams(text)
for conversion in conversions:
    print(f"{conversion[0]} {conversion[1]} = {conversion[2]} gr")
```

```python
import re
from fractions import Fraction

def convert_to_grams(text):
    conversions = {
        "sdm": 12,
        "sdt": 4,
        "ml": 1,
        "suing": 5,
        "siung": 5,
        "butir": 50,
        "buah": 150,
        "daun": 1,
        "potong": 100,
        "bh": 50,
        "bt": 50,
        "biji": 50,
        "bj": 50,
        "sendok makan": 12,
        "sendok teh": 4,
        "btr": 50,
        "ptng": 100,
        "btg": 100,
        "lmbr": 5,
        "cup": 230,
        "gelas": 250,
        "gls": 250,
        "mangkok": 300,
        "secukupnya": 5,
        "gr": 1,
    }


    result = []
    pattern = r'(\d+\s*(?:\d+/\d+)?)(?:\s*(\S+))?\s*\[?(\S*)\]?\s*'
    for line in text.split('\n'):
        for match in re.finditer(pattern, line):
            amount, unit, ingredient = match.groups()
            amount = amount.replace(' ', '')
            amount = eval(amount) if '/' in amount else float(amount)

            if unit in conversions:
                grams = amount * conversions[unit]
                result.append((line, amount, unit, grams, ingredient))
            else:
                result.append((line, None, None, None, None))

    return result

user_input = '''
1 buah [tahu cina]
2 buah [tahu cina]
0.5 buah [tahu cina]
1/2 buah [tahu cina]
1/4 sdt [garam]
2 [siung bawang putih], haluskan
100 ml air
3 buah lontong, potong
150 gram taoge
50 gram bihun
1 sdm kecap manis untuk penyajian
1 sdm bawang merah goreng untuk taburan
50 gram kerupuk kanji, goreng
1 buah mentimun, potong
4 bagian, iris Minyak goreng
'''

user_input = user_input.strip()
conversions = convert_to_grams(user_input)
for conversion in conversions:
    print(f"{conversion[0]} {conversion[1]} = {conversion[2]} gr \tIngredient: {conversion[3]}")
```
